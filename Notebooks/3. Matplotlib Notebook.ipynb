<h3> Display your plot </h3>
    
__`plt.show()`__ starts an event loop, looks for all currently active figure objects,
and opens one or more interactive windows that display your figure or figures

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

<h3> Two Interfaces for the Price of One </h3>

```python
plt.figure() # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1)     # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

<h3> Object-oriented interface </h3>

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

<h3> Adjusting the Plot: Axes Limits </h3>

- The most basic way to adjust axis limits is to use the `plt.xlim()` and `plt.ylim()` methods
- If for some reason youd like either axis to be displayed in reverse, you can simply
reverse the order of the arguments

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 12)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))

# reverse the order to reverse the figure
plt.xlim(12, -1)
plt.ylim(1.5, -1.5)
```

### `plt.axis()`
- The `plt.axis()` method allows you to set the x
and y limits with a single call, by passing a list that specifies [xmin, xmax, ymin,ymax]
- allowing you to do things like automatically
tighten the bounds around the current plot

that on your screen, one unit in x is equal to one unit in y

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal'); # one unit per x = one unit per y
```

<h2 id='S3'> Simple Scatter Plots </h2>

Another commonly used plot type is the simple scatter plot, a close cousin of the line
plot. Instead of points being joined by line segments, here the points are represented
individually with a dot, circle, or other shape.

### When to use it ?
- When you have paired numerical data
- When your dependent variable may have multiple values for each value of your independent variable
- When trying to determine whether the two variables are related

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black') # using .plot()
```

```python
plt.scatter(x, y) # using plt.scatter()
```

The primary difference of `plt.scatter()` from `plt.plot()` is that it can be used to create
scatter plots where the properties of each individual point (size, face color, edge color,
etc.) can be individually controlled or mapped to data.

```python
# Create a random scatter plot
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
s=100*features[3], c=iris.target, cmap='viridis')
```

<h2 id='S4'> Visualizing Errors </h2>

Accurate accounting for errors is nearly as important, if not more important, than accurate reporting of the number itself.

### Basic Errorbars

A basic errorbar can be created with a single Matplotlib function call

```python
dy = 0.3

x = np.linspace(0, 10, 50)
y = np.sin(x) + dy * np.random.randn(50)


# fmt is a format code controlling the appearance of lines and points
plt.errorbar(x, y, yerr=dy, fmt='.k');
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',ecolor='lightgray', elinewidth=3, capsize=0);
```

### Continuous Errors
In some situations it is desirable to show errorbars on continuous quantities. Though
Matplotlib does not have a built-in convenience routine for this type of application,
its relatively easy to combine primitives like `plt.plot()` and `plt.fill_between()` for a
useful result.

```python
# define the model and draw some data

y_lower = [i - .3 * i for i in y]
y_upper = [i + .3 * i for i in y]

plt.fill_between(x, y_lower, y_upper, alpha=.2, color='gray')
plt.plot(x, y, 'k', alpha=.6)
```

<h2 id='S5'> Density and Contour Plots </h2>

Sometimes it is useful to display three-dimensional data in two dimensions using
contours or color-coded regions. There are three Matplotlib functions that can be
helpful for this task: plt.contour for contour plots, plt.contourf for filled contour
plots, and plt.imshow for showing images.

### Visualizing a Three-Dimensional Function

A contour plot can be created with the plt.contour function. It takes three arguments:
a grid of x values, a grid of y values, and a grid of z values. The x and y values
represent positions on the plot, and the z values will be represented by the contour
levels. Perhaps the most straightforward way to prepare such data is to use the
np.meshgrid function, which builds two-dimensional grids from one-dimensional
arrays

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y) # builds two-dimensional grids from one-dimensional arrays
Z = f(X, Y)

plt.contour(X, Y, Z)
```

```python
plt.contour(X, Y, Z, 20, cmap = 'RdGy') # X Y Z number of lines
```

```python
plt.contourf(X, Y, Z, 20, cmap = 'RdGy') # X Y Z number of lines
plt.colorbar()
```

<h2 id='S6'>Histograms, Binnings, and Density</h2>

A simple histogram can be a great first step in understanding a dataset.
When you want to see the shape of the data's distribution,

```python
data = np.random.randn(1000) # create a random data
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=20, edgecolor='w')
plt.show()
```

```python
plt.hist(data, density=True, edgecolor='w')
plt.show()
```

```python
# create some random data

x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40) # parameter we will use

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)

plt.show()
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

<h2 id='S6'>Customizing Plot Legends </h2>

Plot legends give meaning to a visualization, assigning labels to the various plot elements

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')

ax.axis('equal')   # unit of x = unit of y

leg = ax.legend(); # set legend
```

```python
ax.legend(loc='upper left', ncol=1, shadow=True, borderpad=1)
fig
```

```python
# set x, y
x = np.linspace(0, 10, 1000)
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))

lines = plt.plot(x, y)

# Customizing legend elements
plt.legend(lines[:3], ['first', 'second', 'third']);
```

<h2 id='S8'> Multiple Subplots </h2>

subplots: groups of smaller axes that can exist together
within a single figure. These subplots might be insets, grids of plots, or other more
complicated layouts.

<h3> plt.axes: Subplots by Hand </h3>

```python
ax1 = plt.axes() # standard axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
# Object oriented interface
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4])
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4])

x = np.linspace(0, 10)

ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

<h3> plt.subplot: Simple Grids of Subplots </h3>

```python
plt.figure(figsize=(12,5))
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
```

<h3> plt.subplots: The Whole Grid in One Go </h3>

well create a 23 grid of subplots, where all axes in the same row share their
y-axis scale, and all axes in the same column share their x-axis scale

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
ax[0, 2].plot(x, np.sin(x))
ax[1, 1].plot(x, np.cos(x))
```

<h3> plt.GridSpec: More Complicated Arrangements </h3>

```python
# we can specify subplot locations and extents using the familiar Python slicing syntax

grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```
